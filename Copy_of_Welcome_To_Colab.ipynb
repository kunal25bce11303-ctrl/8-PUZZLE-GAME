```
import heapq
from copy import deepcopy

# 1. FUNCTIONS FIRST (before input)
def is_valid_puzzle(puzzle):
    flat = [num for row in puzzle for num in row]
    return sorted(flat) == list(range(9))

def is_solved(current, GOAL):
    return current == GOAL

def manhattan_distance(puzzle, GOAL):
    distance = 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] != 0:
                # Find this tile's goal position
                for gi in range(3):
                    for gj in range(3):
                        if GOAL[gi][gj] == puzzle[i][j]:
                            distance += abs(i-gi) + abs(j-gj)
                            break
    return distance

def solve_8puzzle(start_puzzle, GOAL):
    open_list = []
    heapq.heappush(open_list, (0, 0, start_puzzle, []))
    closed = set()

    while open_list:
        f, g, current, path = heapq.heappop(open_list)
        flat = tuple(tuple(row) for row in current)
        if flat in closed: continue
        closed.add(flat)

        if is_solved(current, GOAL):
            return path + [current], len(closed)

        blank_r, blank_c = next((r,c) for r in range(3) for c in range(3) if current[r][c] == 0)

        for dr, dc, move in [(-1,0,'U'), (1,0,'D'), (0,-1,'L'), (0,1,'R')]:
            nr, nc = blank_r + dr, blank_c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                next_state = deepcopy(current)
                next_state[blank_r][blank_c], next_state[nr][nc] = next_state[nr][nc], next_state[blank_r][blank_c]

                if tuple(tuple(row) for row in next_state) not in closed:
                    h = manhattan_distance(next_state, GOAL)
                    heapq.heappush(open_list, (g+1+h, g+1, next_state, path + [move]))

    return None, len(closed)

# 2. INPUT + VALIDATION
print("Enter START puzzle (0-8 exactly once):")
a,b,c = map(int, input("Row 1: ").split())
d,e,f = map(int, input("Row 2: ").split())
g,h,i = map(int, input("Row 3: ").split())
puzzle = [[a,b,c],[d,e,f],[g,h,i]]

print("Enter GOAL puzzle (0-8 exactly once):")
j,k,l = map(int, input("Goal Row 1: ").split())
m,n,o = map(int, input("Goal Row 2: ").split())
p,q,r = map(int, input("Goal Row 3: ").split())
GOAL = [[j,k,l],[m,n,o],[p,q,r]]

# 3. VALIDATE
if not is_valid_puzzle(puzzle) or not is_valid_puzzle(GOAL):
    print(" Invalid puzzle(s)! Must contain 0-8 exactly once.")
    exit()

print(f"\nStart:\n{puzzle}")
print(f"Goal:\n{GOAL}")

# 4. SOLVE
result, nodes = solve_8puzzle(puzzle, GOAL)
if result:
    print(f"\n SOLVED in {len(result)-1} moves, {nodes} nodes explored!")
    for i, state in enumerate(result):
        print(f"Step {i}: {state}")
else:
    print(f"\n Unsolvable (explored {nodes} nodes)")
def is_solvable(start, goal):
    def inversions(puzzle):
        flat = [x for row in puzzle for x in row if x != 0]
        count = 0
        for i in range(len(flat)):
            for j in range(i+1, len(flat)):
                if flat[i] > flat[j]:
                    count += 1
        return count

    return inversions(start) % 2 == inversions(goal) % 2

# Before solve_8puzzle():
if not is_solvable(puzzle, GOAL):
    print(" UNSOLVABLE - different parity spaces!")
    exit()
print(" READY - A* Algorithm Complete!")
input("Press Enter to exit...")
```
